```python
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
import datetime
import warnings


from scipy import stats
from itertools import product
```

```python
data = pd.read_csv('kshen.txt', ';')
```

```python
data
```

```python
def parser(year, day):
    return datetime.datetime(int(year), 1, 1) + datetime.timedelta(int(day) - 1)
```

```python
data['date'] = data.apply(lambda x: parser(x['year'], x['day']), axis=1)
```

```python
def data_parser(dataframe, column, idx):
    tmp = dataframe[['date', column]]
    tmp = tmp.groupby(pd.Grouper(key='date', axis=0, freq=f'{idx}')).mean()
    tmp.to_csv(f'data/{column}/full_{idx}.csv')
    tmp[tmp.index.year != 2015].to_csv(f'data/{column}/train_{idx}.csv')
    tmp[tmp.index.year == 2015].to_csv(f'data/{column}/test_{idx}.csv')
```

```python
days = ['1d', '5d', '10d', '15d', 'M']
columns = ['tavg', 'tmin', 'tmax', 'wind', 'sunhours', 'globrad', 'precip', 'relhumid']
for day in days:
    for col in columns:
        data_parser(data, col, day)
```
