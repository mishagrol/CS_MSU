```python
from interval import Interval
import random

import matplotlib.pyplot as plt 
```

```python
%matplotlib inline
plt.rcParams["figure.figsize"] = (12, 9)
```

```python
def rand(x):
    return x.a + (x.b - x.a) * random.random()
```

```python
def rand_r(x):
    return random.choice([x.a, x.b])
```

```python
def zero_matrix(n):
    return [([0]*n) for i in range(n)]

def zero_vector(n):
    return [0 for i in range(n)]
```

```python
def gauss(a, b, n):
    _a = zero_matrix(n)
    _b = zero_vector(n)
    x = zero_vector(n)
    
    for i in range(0, n):
        for j in range(0, n):
            _a[i][j] = a[i][j]
        _b[i] = b[i]
        
    for i in range(0, n):
        for j in range(i+1, n):
            _b[j] = _b[j] - _b[i] * _a[j][i] / _a[i][i]
            for k in range(i+1, n):
                _a[j][k] = _a[j][k] - _a[i][k] * _a[j][i] / _a[i][i]
    
    for j in range(n-1, -1, -1):
        x[j] = _b[j]
        for k in range(j+1, n):
            x[j] = x[j]  - _a[j][k] * x[k];
        x[j] = x[j] / _a[j][j];
    
    return x
```
